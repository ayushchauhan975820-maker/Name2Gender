```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df = df[['Name', 'Gender']].copy()
df['Gender'] = df['Gender'].map({'M': 1, 'F': 0})
df['Gender'] = df['Gender'].astype(int)
```

```python
df.head()
```

```python
X = df
y = df["Gender"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42, shuffle = True)
```

```python
from Model_Naive_Bayes import GenderNaiveBayes

model = GenderNaiveBayes()
print("Training Model")
model.fit(X_train)
print("Training Complete")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

print("prediction")

predictions_df = model.predict(X_test["Name"])
predictions_df.head()
```

```python
y_pred = predictions_df[0] 

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy:  {accuracy * 100:.2f}%")
print(f"Precision: {precision * 100:.2f}%")
print(f"Recall:    {recall * 100:.2f}%")
print(f"F1-Score:  {f1 * 100:.2f}%")

print("\n Classification Report")
print(classification_report(y_test, y_pred, target_names=["Female (0)", "Male (1)"]))
```
